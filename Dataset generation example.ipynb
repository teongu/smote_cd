# Generation of the synthetic dataset

### Example of generation with 2 features and 2 labels

To generate a synthetic dataset, we first need to generate a $\beta$ matrix. If $n$ is the number of features and $K$ the number of classes, then $\beta$ will have a size $n \times (K+1)$. In this example, we select $n=2$ and $K=2$

```python
n_features = 2
n_classes = 2
betas = dataset_generation.generate_betas(n_features,n_classes,random_state=0)
print(betas)
```

Once we have the $\beta$ matrix, we select the size of the dataset and we can generate the features and the labels. The features are generated in $[-10,10]^n$.

```python
size = 1000
generated_X,generated_y,_ = dataset_generation.generate_dataset(n_features,n_classes,size,betas=betas,random_state=0)
```

With two features, we can plot all the points, each axis being a feature. The color of the points can be defined by one of the two classes.

```python
plt.scatter(generated_X[:,0],generated_X[:,1],c=generated_y[:,0])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### Effect of the $\beta$ matrix

We see here the differences that can be induced by the $\beta$ matrix. We generate 3 different $\beta$ matrices and create three different datasets with them. The features remain the same, thanks to the `random_state=0` argument.

```python
betas_1 = dataset_generation.generate_betas(n_features,n_classes,random_state=1)
betas_2 = dataset_generation.generate_betas(n_features,n_classes,random_state=2)
betas_3 = dataset_generation.generate_betas(n_features,n_classes,random_state=3)

generated_X_1,generated_y_1,_ = dataset_generation.generate_dataset(n_features,n_classes,size,betas=betas_1,random_state=0)
generated_X_2,generated_y_2,_ = dataset_generation.generate_dataset(n_features,n_classes,size,betas=betas_2,random_state=0)
generated_X_3,generated_y_3,_ = dataset_generation.generate_dataset(n_features,n_classes,size,betas=betas_3,random_state=0)
```

```python
fig,ax = plt.subplots(1,3,figsize=(12,6))
ax[0].scatter(generated_X_1[:,0],generated_X_1[:,1],c=generated_y_1[:,0],s=10)
ax[1].scatter(generated_X_2[:,0],generated_X_2[:,1],c=generated_y_2[:,0],s=10)
ax[2].scatter(generated_X_3[:,0],generated_X_3[:,1],c=generated_y_3[:,0],s=10)
plt.show()
```
