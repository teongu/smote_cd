```python
import numpy as np
import matplotlib.pyplot as plt
import dataset_generation
import smote_cd
```

# Generation of the data

We first generate a synthetic dataset with a well-chosen $\beta$ matrix so that the dataset will be imbalanced. For more information about the generation of the datasets, you can have a look at the notebook *Dataset generation example.ipynb*.

```python
n_features = 2
n_classes = 2
size = 1000

betas = [[0.8, 0.8, 0.8],[-3, 0.2, 0.6]]
generated_X,generated_y,_ = dataset_generation.generate_dataset(n_features,n_classes,size,betas=betas,random_state=0)
```

```python
plt.scatter(generated_X[:,0],generated_X[:,1],c=generated_y[:,0],s=10)
plt.show()
```

Almost 80% of the points are from class 0 :

```python
sum(generated_y)
```
